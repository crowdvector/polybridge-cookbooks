This notebook uses the local `stress_monitor.py` helper when it is present and falls back to the final GitHub path after `crowdvector/polybridge-cookbooks` exists and has been pushed.

Outputs are market-implied snapshots from live prediction markets and can change as source markets update.

```python
import importlib.util
import pathlib
import urllib.request

from IPython.display import Image, display

HELPER_PATH = pathlib.Path("stress_monitor.py")
REMOTE_HELPER_URL = "https://raw.githubusercontent.com/crowdvector/polybridge-cookbooks/main/vix-forecast/stress_monitor.py"

if not HELPER_PATH.exists():
    urllib.request.urlretrieve(REMOTE_HELPER_URL, HELPER_PATH)

spec = importlib.util.spec_from_file_location("stress_monitor", HELPER_PATH)
stress_monitor = importlib.util.module_from_spec(spec)
assert spec.loader is not None
spec.loader.exec_module(stress_monitor)
```

```python
snapshot_path, chart_path, snapshot = stress_monitor.run_once(prompt_for_key=True)
stress_monitor.print_summary(snapshot)
display(Image(filename=str(chart_path)))
```
